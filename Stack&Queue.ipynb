```python
# Function to print the puzzle
def print_puzzle(puzzle):
    for row in puzzle:
        print(' '.join(map(str, row)))

# Function to find the coordinates of a number in the puzzle
def find_coordinates(puzzle, num):
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == num:
                return i, j

# Function to calculate the distance heuristic
def calculate_distance(puzzle):
    distance = 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] != 0:
                x_goal, y_goal = divmod(puzzle[i][j] - 1, 3)
                distance += abs(i - x_goal) + abs(j - y_goal)
    return distance

# Function to generate possible moves for the blank tile
def generate_moves(puzzle):
    moves = []
    x_blank, y_blank = find_coordinates(puzzle, 0)
    if x_blank > 0:
        moves.append((x_blank - 1, y_blank))  # Move Up
    if x_blank < 2:
        moves.append((x_blank + 1, y_blank))  # Move Down
    if y_blank > 0:
        moves.append((x_blank, y_blank - 1))  # Move Left
    if y_blank < 2:
        moves.append((x_blank, y_blank + 1))  # Move Right
    return moves

# Function to perform the A* search algorithm
def solve_puzzle(initial_state):
    open_list = [(calculate_distance(initial_state), 0, initial_state)]
    closed_list = set()

    while open_list:
        open_list.sort()
        _, moves, current_state = open_list.pop(0)
        closed_list.add(tuple(map(tuple, current_state)))

        if current_state == [[1, 2, 3], [4, 5, 6], [7, 8, 0]]:
            return current_state, moves

        possible_moves = generate_moves(current_state)
        x_blank, y_blank = find_coordinates(current_state, 0)

        for move in possible_moves:
            new_state = [list(row) for row in current_state]
            x, y = move
            new_state[x_blank][y_blank], new_state[x][y] = new_state[x][y], new_state[x_blank][y_blank]

            if tuple(map(tuple, new_state)) not in closed_list:
                open_list.append((moves + 1 + calculate_distance(new_state), moves + 1, new_state))

# Example usage:
initial_state = [[1, 2, 3], [4, 5, 0], [6, 7, 8]]  # Initial state of the puzzle
print("Initial state:")
print_puzzle(initial_state)
solution, moves = solve_puzzle(initial_state)
print("Solution after", moves, "moves:")
print_puzzle(solution)
```

```python
# Function to take user input for a matrix
def take_input():
    print("Enter the 3x3 matrix (Enter 0 for blank space):")
    matrix = []
    for _ in range(3):
        row = list(map(int, input().split()))
        matrix.append(row)
    return matrix

# Function to take user input for the current and goal state
def input_states():
    print("Enter the initial state:")
    initial_state = take_input()
    print("Enter the goal state:")
    goal_state = take_input()
    return initial_state, goal_state

# Function to print all the matrices showing the moves taken
def print_solution_matrices(solution_matrices):
    for idx, matrix in enumerate(solution_matrices):
        print(f"Move {idx + 1}:")
        print_puzzle(matrix)
        print()

# Function to calculate the distance heuristic
def calculate_distance(puzzle):
    distance = 0
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] != 0:
                x_goal, y_goal = divmod(puzzle[i][j] - 1, 3)
                distance += abs(i - x_goal) + abs(j - y_goal)
    return distance

# Function to find the coordinates of a number in the puzzle
def find_coordinates(puzzle, num):
    for i in range(3):
        for j in range(3):
            if puzzle[i][j] == num:
                return i, j

# Function to generate possible moves for the blank tile
def generate_moves(puzzle):
    moves = []
    x_blank, y_blank = find_coordinates(puzzle, 0)
    if x_blank > 0:
        moves.append((x_blank - 1, y_blank))  # Move Up
    if x_blank < 2:
        moves.append((x_blank + 1, y_blank))  # Move Down
    if y_blank > 0:
        moves.append((x_blank, y_blank - 1))  # Move Left
    if y_blank < 2:
        moves.append((x_blank, y_blank + 1))  # Move Right
    return moves

# Function to solve the puzzle and track the solution matrices
def solve_puzzle_with_tracking(initial_state, goal_state):
    open_list = [(calculate_distance(initial_state), 0, initial_state)]"""This line removes and returns the first element (the one with the lowest cost) from the sorted open_list.
          The underscore _ is used to discard the first value in the tuple returned by pop(0). In this context, it's typically used as a placeholder for a value that isn't needed.
          moves represents the number of moves made to reach the current state.
          current_state represents the state of the puzzle/grid after applying the moves."""
    closed_list = set()#This line converts the current state, which is a list of lists,
        #into a tuple of tuples to make it hashable (as lists are not hashable)
        #and adds this tuple representation of the current state to the closed_list.
    solution_matrices = [initial_state]

    while open_list:
        open_list.sort()
        _, moves, current_state = open_list.pop(0)
        closed_list.add(tuple(map(tuple, current_state)))

        if current_state == goal_state:
            return solution_matrices, moves

        possible_moves = generate_moves(current_state)
        x_blank, y_blank = find_coordinates(current_state, 0)

        for move in possible_moves:
            new_state = [list(row) for row in current_state]
            #creates a copy of the current state of the puzzle
            x, y = move
            new_state[x_blank][y_blank], new_state[x][y] = new_state[x][y], new_state[x_blank][y_blank]

            if tuple(map(tuple, new_state)) not in closed_list:
                open_list.append((moves + 1 + calculate_distance(new_state), moves + 1, new_state))
                solution_matrices.append(new_state)

if __name__ == "__main__": #Main Program
    initial_state, goal_state = input_states()

    solution_matrices, moves = solve_puzzle_with_tracking(initial_state, goal_state)

    print("\nSolution:")
    print_solution_matrices(solution_matrices)
    print("Solution found after", moves, "states.")
```
